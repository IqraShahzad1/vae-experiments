```lua
require 'hdf5';
```

```lua
f = hdf5.open('cifar10.hdf5', 'r')
train = f:read('xTrain'):all():double()
valid = f:read('xTest'):all():double()
ytrain = f:read('yTrain'):all():double()
yvalid = f:read('yTest'):all():double()
f:close()
```

```lua
-- Div by 255
train:div(255)
valid:div(255)
print(train:max())
print(valid:max())
```

```lua
myFile = hdf5.open('cifar.hdf5', 'w')
```

```lua
options = hdf5.DataSetOptions()
options:setChunked(1024)
options:setDeflate()
```

```lua
myFile:write('xTrain', train, options)
```

```lua
myFile:write('xTest', valid, options)
```

```lua
myFile:write('yTrain', ytrain, options)
myFile:write('yTest', yvalid, options)
```

```lua
myFile:close()
```
